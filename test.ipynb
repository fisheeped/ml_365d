```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

```python
train_images, test_images = train_images / 255.0, test_images / 255.0

train_images.shape,test_images.shape,train_labels.shape,test_labels.shape
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
data_enrichment = ImageDataGenerator(
    rescale=1. / 1.,  # 如为图像，分母为255时，可归至0～1
    rotation_range=45,  # 随机45度旋转
    width_shift_range=.05,  # 宽度偏移
    height_shift_range=.05,  # 高度偏移
    horizontal_flip=False,  # 水平翻转
    zoom_range=.5  # 将图像随机缩放阈量50％
)
```

```python

fig = plt.figure(figsize=(20, 2))
for x_batch in data_enrichment.flow(train_images, batch_size=12, shuffle=False):
    for i in range(0, 12):
        ax = fig.add_subplot(1, 12, i + 1)
        ax.imshow(x_batch[i])
    fig.suptitle('Augmented Images', fontsize=20)# 总标题
    plt.show()
    break
```
