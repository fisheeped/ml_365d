```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt

(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()
```

# 数据处理
## 数据归一化以及数据维度信息
- (50000, 32, 32, 3)指的是5000张32x32的图片，每个像素为(0~255)*3  
- 50000份训练集较少，考虑增加训练集

```python
train_images, test_images = train_images / 255.0, test_images / 255.0

train_images.shape,test_images.shape,train_labels.shape,test_labels.shape
```

## 可视化图片
10列形式纵向叠加展示图片

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer','dog', 'frog', 'horse', 'ship', 'truck']
plt.figure(figsize=(20,10))
for i in range(20):
    plt.subplot(4,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()
```

## 模型构建
参考(https://www.heywhale.com/mw/project/62e72619d9ad8bab8167f50c)

```python
model = models.Sequential([
    layers.Conv2D(32, (3, 3),padding='same', activation='relu', input_shape=(32, 32, 3)), #卷积层1，卷积核3*3
    layers.MaxPooling2D((2, 2)),   #池化层1，2*2采样

    layers.Conv2D(64, (3, 3),padding='same', activation='relu'), #卷积层2，卷积核3*3
    layers .MaxPooling2D((2, 2)),   #池化层2，2*2采样

    layers.Conv2D(64, (3, 3),padding='same', activation='relu'),  #卷积层3，卷积核3*3
    layers.MaxPooling2D((2, 2)),     

    layers.Conv2D(64, (3, 3),padding='same', activation='relu'),  #卷积层4，卷积核3*3
    layers.Dropout(0.1),
    layers.MaxPooling2D((2, 2)),                #池化层4，2*2采样

    layers.Conv2D(64, (3, 3),padding='same', activation='relu'),  #卷积层5，卷积核3*3
    layers.Dropout(0.1),
    
    layers.Flatten(),                      #Flatten层，连接卷积层与全连接层
    layers.Dense(64, activation='relu'),   #全连接层，特征进一步提取
    layers.Dense(10,activation="softmax")                       #输出层，输出预期结果
])

model.summary()  # 打印网络结构
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr = 0.001),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(train_images, train_labels, epochs=20,batch_size = 512, 
                    validation_data=(test_images, test_labels))
```

## 数据增强
- 数据增强，并将增强数据集加入训练集  
- 通俗来说就是图片拉扯，旋转，缩放，风格化

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
data_enrichment = ImageDataGenerator(
    rescale=1. / 1.,  # 如为图像，分母为255时，可归至0～1
    rotation_range=45,  # 随机45度旋转
    width_shift_range=.15,  # 宽度偏移
    height_shift_range=.15,  # 高度偏移
    horizontal_flip=False,  # 水平翻转
    zoom_range=0.2  # 将图像随机缩放阈量20％
)
```

```python
history = model.fit(data_enrichment.flow(train_images, train_labels,batch_size=32), 
                    epochs=20,
                    validation_data=(test_images, test_labels), 
                    validation_freq=4)
```

### 多次增强数据

```python
data_enrichment = ImageDataGenerator(
    rescale=1. / 1.,  # 如为图像，分母为255时，可归至0～1
    rotation_range=30,  # 随机30度旋转
    width_shift_range=.05,  # 宽度偏移
    height_shift_range=.05,  # 高度偏移
    horizontal_flip=False,  # 水平翻转
    zoom_range=0.2  # 将图像随机缩放阈量20％
)
history = model.fit(data_enrichment.flow(train_images, train_labels,batch_size=32), 
                    epochs=10,
                    validation_data=(test_images, test_labels), 
                    validation_freq=2)
```

```python
history = model.fit(data_enrichment.flow(train_images, train_labels,batch_size=32), 
                    epochs=20,
                    validation_data=(test_images, test_labels), 
                    validation_freq=2)
```

```python
data_enrichment = ImageDataGenerator(
    rescale=1. / 1.,  # 如为图像，分母为255时，可归至0～1
    rotation_range=15,  # 随机15度旋转
    width_shift_range=.10,  # 宽度偏移
    height_shift_range=.10,  # 高度偏移
    horizontal_flip=False,  # 水平翻转
    zoom_range=0.1  # 将图像随机缩放阈量20％
)
history = model.fit(data_enrichment.flow(train_images, train_labels,batch_size=32), 
                    epochs=20,
                    validation_data=(test_images, test_labels), 
                    validation_freq=2)
```

```python
import numpy as np

pre = model.predict(test_images)
print(class_names[np.argmax(pre[1])])
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='accuracy')
plt.plot(history.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
plt.show()

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
```
